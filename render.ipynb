```python
from smash_rl.video import dump_replay_video, extract_frame_records, compose_video, write_calibration_video, specs_for_replay
import ipywidgets as widgets
import os
import smash_rl.specs as specs
```

```python
available_dirs = os.listdir("replays")
dir_selector = widgets.Dropdown(
    options=available_dirs,
    description='Select Directory:',
    disabled=False,
)
display(dir_selector)
```

```python
instances = [f for f in os.listdir(f"replays/{dir_selector.value}") if os.path.isdir(f"replays/{dir_selector.value}/{f}")]

available_replays = []
for instance in instances:
    replays = [f for f in os.listdir(f"replays/{dir_selector.value}/{instance}") if f.endswith(".slp")]
    available_replays.extend([f"{instance}/{replay}" for replay in replays])

replay_selector = widgets.Dropdown(
    options=sorted(available_replays, reverse=True),
    description='Select Replay:',
    disabled=False,
)
display(replay_selector)
```

```python
available_models = [f for f in os.listdir("weights") if f.endswith(".zip")]
model_selector = widgets.Dropdown(
    options=available_models,
    description='Select Model:',
    disabled=False,
)
display(model_selector)
```

```python
# Spec di obs/azioni per la diagnostica. "(auto)" = leggile dal config.json della run
# che ha prodotto il replay (replays/<run>/... -> runs/<run>/config.json); scegli un
# valore esplicito per forzarlo manualmente.
obs_spec_selector = widgets.Dropdown(
    options=["(auto)"] + list(specs.OBS_SPECS),
    description='Select Obs Spec:',
    disabled=False,
)
act_spec_selector = widgets.Dropdown(
    options=["(auto)"] + list(specs.ACT_SPECS),
    description='Select Act Spec:',
    disabled=False,
)
display(obs_spec_selector, act_spec_selector)
```

```python
slp = f"replays/{dir_selector.value}/{replay_selector.value}"
model = f"weights/{model_selector.value}"

# spec: "(auto)" -> dal config.json della run, altrimenti il valore scelto nei dropdown
auto_obs, auto_act = specs_for_replay(slp)
obs_name = auto_obs if obs_spec_selector.value == "(auto)" else obs_spec_selector.value
act_name = auto_act if act_spec_selector.value == "(auto)" else act_spec_selector.value
if obs_name is None or act_name is None:
    raise ValueError(f"spec non derivabili da {slp} (obs={obs_name}, act={act_name}): "
                     "seleziona obs/act manualmente nei dropdown")
print(f"obs_spec={obs_name}  act_spec={act_name}")

dump = dump_replay_video(slp)                        # .slp -> mp4 (lento: tempo reale)
records = extract_frame_records(slp, model, obs_name=obs_name, act_name=act_name)   # diagnostica per-frame (veloce, senza Dolphin)
compose_video(dump, records, f"videos/annotated_{replay_selector.value.split('/')[-1]}.mp4",
              obs_name=obs_name, act_name=act_name)   # gameplay + pannello
```
