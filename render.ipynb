```python
from smash_rl.video import dump_replay_video, extract_frame_records, compose_video, write_calibration_video
import ipywidgets as widgets
import os
import smash_rl.specs as specs
```

```python
available_dirs = os.listdir("replays")
dir_selector = widgets.Dropdown(
    options=available_dirs,
    description='Select Directory:',
    disabled=False,
)
display(dir_selector)
```

```python
instances = [f for f in os.listdir(f"replays/{dir_selector.value}") if os.path.isdir(f"replays/{dir_selector.value}/{f}")]

available_replays = []
for instance in instances:
    replays = [f for f in os.listdir(f"replays/{dir_selector.value}/{instance}") if f.endswith(".slp")]
    available_replays.extend([f"{instance}/{replay}" for replay in replays])

replay_selector = widgets.Dropdown(
    options=available_replays,
    description='Select Replay:',
    disabled=False,
)
display(replay_selector)
```

```python
available_models = [f for f in os.listdir("weights") if f.endswith(".zip")]
model_selector = widgets.Dropdown(
    options=available_models,
    description='Select Model:',
    disabled=False,
)
display(model_selector)
```

```python
available_obs_specs = [f for f in specs.OBS_SPECS]
obs_spec_selector = widgets.Dropdown(
    options=available_obs_specs,
    description='Select Obs Spec:',
    disabled=False,
)
display(obs_spec_selector)
```

```python
slp = f"replays/{dir_selector.value}/{replay_selector.value}"
model = f"weights/{model_selector.value}"

dump = dump_replay_video(slp)                        # .slp -> mp4 (lento: tempo reale)
records = extract_frame_records(slp, model, obs_name=obs_spec_selector.value)          # diagnostica per-frame (veloce, senza Dolphin)
compose_video(dump, records, f"videos/annotated_{replay_selector.value.split('/')[-1]}.mp4")        # gameplay + pannello
```
